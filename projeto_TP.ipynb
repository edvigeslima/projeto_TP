```python
#Juntando os valores de ORGANIZATION_TYPE: Passo 2 - Tratar os dados através de um apply e replace e Atribuir à coluna

def tratar_ORGANIZATION_TYPE(x):
    for i in range(14):
        x = str(x).replace(' Type '+str(i),'').replace(': type '+(str(i)) if (i < 10) else str(i-10),'')
    return x

df_home_credit_default_risk['ORGANIZATION_TYPE'] = df_home_credit_default_risk['ORGANIZATION_TYPE'].apply(lambda x: tratar_ORGANIZATION_TYPE(x))
```

```python
# precisa tratar a coluna 'DAYS_BIRTH', inserir a nova coluna 'AGE_BIRTH'
df_home_credit_default_risk['AGE_BIRTH'] = df_home_credit_default_risk['DAYS_BIRTH'].apply(lambda x: int(-x/365))
```

# 01) Qual é a idade, em anos, do titular mais jovem?

```python
youngest = df_home_credit_default_risk['AGE_BIRTH'].min()
print(f'O titular mais jovem tem {youngest} anos.')
```

# 02) Qual a média do valor de anuidade dos residentes em apartamentos maiores de 50 anos?

```python
filter = (df_home_credit_default_risk['AGE_BIRTH']>50) & (df_home_credit_default_risk['HOUSETYPE_MODE'] == 'block of flats')
mean_50years = df_home_credit_default_risk[filter]['AMT_ANNUITY'].mean()

print(f'A média do valor de anuidade dos residentes em apartamentos maiores de 50 anos é R$ {mean_50years:.2f}')
```

# 03) Existe uma correlação forte entre ter imóvel próprio e ter carro próprio? Por que?

```python
# Tratando a coluna 'FLAG_OWN_CAR' atribuindo 1 para sim e 0 para não
df_home_credit_default_risk['FLAG_OWN_CAR'] = df_home_credit_default_risk['FLAG_OWN_CAR'].apply(lambda x: 1 if x == 'Y' else 0)
df_home_credit_default_risk['FLAG_OWN_CAR']
```

```python
# Tratando a coluna 'FLAG_OWN_REALTY' atribuindo 1 para sim e 0 para não
df_home_credit_default_risk['FLAG_OWN_REALTY'] = df_home_credit_default_risk['FLAG_OWN_REALTY'].apply(lambda x: 1 if x == 'Y' else 0)
df_home_credit_default_risk['FLAG_OWN_REALTY']
```

```python
# a correlação entre ter imóvel próprio e ter carro próprio
df_home_credit_default_risk[['FLAG_OWN_REALTY', 'FLAG_OWN_CAR']].corr()
```

```python
# Por quê?
# A correlação entre ter imóvel próprio e ter carro próprio indica uma correlação positivo fraca, 
# porque um coeficiente de correlação próximo de zero indica que não há relação entre as duas variáveis,
# e quanto mais eles se aproximam de 1 ou -1, mais forte é a relação.
```

# 06) Considerando os tipos de ocupação, podemos dizer que os "Managers" possuem um valor de anuidade mais alto? Se não, qual é?

```python
# Sim, os "Managers" possuem um valor de anuidade mais alto.
df_home_credit_default_risk.groupby('OCCUPATION_TYPE')['AMT_ANNUITY'].max().sort_values(ascending=False)
```

# 08) Qual a média de requisições de empréstimo entre trabalhadores com tipo de ocupação "Sales staff", que não moram com parentes, declararam não estarem em situação de emergência e possuem telefone de trabalho?

```python
filter_1 = (df_home_credit_default_risk['OCCUPATION_TYPE'] == 'Sales staff') & (df_home_credit_default_risk['NAME_HOUSING_TYPE'] != 'With parents') & (df_home_credit_default_risk['EMERGENCYSTATE_MODE'] != 'Yes') & (df_home_credit_default_risk['FLAG_WORK_PHONE'] == 1)
AMT_REQ = ['AMT_REQ_CREDIT_BUREAU_HOUR', 'AMT_REQ_CREDIT_BUREAU_DAY',
       'AMT_REQ_CREDIT_BUREAU_WEEK', 'AMT_REQ_CREDIT_BUREAU_MON',
       'AMT_REQ_CREDIT_BUREAU_QRT', 'AMT_REQ_CREDIT_BUREAU_YEAR']
df_home_credit_default_risk.groupby(filter_1)[AMT_REQ].mean().round(2)
```

# 09) Como você explica a correlação entre o número de filhos e idade (em anos)?

```python
df_home_credit_default_risk[['CNT_CHILDREN','AGE_BIRTH']].corr()
```

```python
# Por quê?
# A correlação entre o número de filhos e idade (em anos) indica uma correlação negativa fraca, 
# porque um coeficiente de correlação próximo de zero indica que não há relação entre as duas variáveis,
# e quanto mais eles se aproximam de 1 ou -1, mais forte é a relação.

# Ou seja, à medida que a idade aumenta, o número de filhos descresce.
```

# 10) Podemos concluir que, quanto mais tempo empregado, maior o crédito? Por quê?

```python
# Passando para anos e tirando os negativos
df_home_credit_default_risk['WORKING_TIME'] = df_home_credit_default_risk['DAYS_EMPLOYED'].apply(lambda x: int(-x/365) if x < 0 else int(x/365))
```

```python
# atribuindo -1 para os que têm 1000 anos de empregados
df_home_credit_default_risk['WORKING_TIME'] = df_home_credit_default_risk['WORKING_TIME'].apply(lambda x: -1 if x > 50 else x)
```

```python
df_home_credit_default_risk[['WORKING_TIME','AMT_CREDIT']].sort_values(by='AMT_CREDIT',ascending=True)
```

```python
df_home_credit_default_risk[['WORKING_TIME','AMT_CREDIT']].max()
```

```python
filtro_2 = (df_home_credit_default_risk['WORKING_TIME']==0)
df_home_credit_default_risk.groupby(filtro_2)['AMT_CREDIT'].max()
```

```python
# Não podemos concluir que "quanto mais tempo empregado, maior o crédito". 
# Pessoas empregadas a menos de um ano possuem mesmo valor de crédito que pessoas empregadas há 48 anos.
# Dessa forma, podemos concluir que outros critérios são analisados para a concessão do crédito.
```

# 11) Qual o desvio padrão da quantidade de filhos de quem tem 8 familiares?
